```python
#Se importan las librerías y se corrobora la ubicación
import csv
import os

print("Current working directory:", os.getcwd())
```

```python
#Se lee el archivo y se visualizan los header
file = open(r'C:\fifa_eda.csv', 'r', encoding='utf-8')

csvreader = csv.reader(file)

header = []
header = next(csvreader)
header
```

```python
#Guarda las filas en una lista
rows = []
for row in csvreader:
    rows.append(row)
```

```python
#Primeras 20 filas
print("Primeras 20 filas:")
print(rows[:20])
```

```python
#Últimas 5 filas
print("Últimas 5 filas:")
print(rows[-5:])
```

```python
#Sample 10
print("Sample de 10 filas:")
print(rows[10:20])
```

```python
#Conversión numérica de Value y Wage
value_idx = header.index('Value')
wage_idx = header.index('Wage')

for row in rows:
    row[value_idx] = row[value_idx].replace('$', '').replace(',', '')
    row[wage_idx] = row[wage_idx].replace('$', '').replace(',', '')

    if row[value_idx] != '':
        # Convert to float first, then to int to handle decimal values
        row[value_idx] = int(float(row[value_idx]))

    if row[wage_idx] != '':
        # Convert to float first, then to int to handle decimal values
        row[wage_idx] = int(float(row[wage_idx]))
```

```python
#Adición de nueva columna
joined_idx = header.index('Joined')
header.append('Years Playing')
```

```python
#Cálculo de años en activo
current_year = 2026

for row in rows:
    years_playing = current_year - int(row[joined_idx])
    row.append(years_playing)
```

```python
#Búsqueda de jugadores en México
nationality_idx = header.index('Nationality')

print("Jugadores de Mexico:")
for row in rows:
    if row[nationality_idx] == 'Mexico':
        print(row)
```

```python
#Ordenamiento
if 'ReleaseClause' in header:
    release_idx = header.index('ReleaseClause')
    
    for row in rows:
        row[release_idx] = row[release_idx].replace('$', '').replace(',', '')
        if row[release_idx] != '':
            row[release_idx] = int(row[release_idx])
else:
    print("'ReleaseClause' column not found in header. Available columns:", header)
```

```python
rows_sorted = sorted(rows, key=lambda x: x[release_idx], reverse=True)

print("Top 10 por ReleaseClause:")
for row in rows_sorted[:10]:
    print(row)
```

```python
players_per_year = {}

for row in rows:
    year = row[joined_idx]
    if year in players_per_year:
        players_per_year[year] += 1
    else:
        players_per_year[year] = 1
```

```python
print("Jugadores por año:")
for year, count in players_per_year.items():
    print(year, count)
```

```python
with open("players_per_year.csv", "w", newline="", encoding="utf-8") as new_file:
    writer = csv.writer(new_file)
    writer.writerow(['Joined', 'Num_Players'])

    for year, count in players_per_year.items():
        writer.writerow([year, count])
```
