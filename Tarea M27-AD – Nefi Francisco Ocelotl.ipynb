```python
# EJERCICIO 1:
# Generar excepciones y mensajes de error correspondientes
filename = "test.txt"
# Intento de abrir un archivo que NO existe
try:
    file = open(filename, "r")
except FileNotFoundError as e:
    print("Error: El archivo no fue encontrado en el directorio.")
    print("Detalle técnico:", e)
except PermissionError as e:
    print("Error: No se tienen permisos para acceder al archivo.")
    print("Detalle técnico:", e)
else:
    print("Archivo abierto correctamente.")
finally:
    print("Finalizando búsqueda de archivo.")
```

```python
# EJERCICIO 2:
# Excepciones y funciones try/except (versión simple)

filename = "test.txt"

try:
    open(filename, "r")
except FileNotFoundError:
    print("No se encontró el archivo:", filename)
```

```python
# EJERCICIO 3:
# Ejemplo con try/except/else/finally

filename = "test.txt"

try:
    archivo = open(filename, "r")
except FileNotFoundError:
    print("Error: No existe el archivo solicitado.")
except Exception as error:
    print("Ocurrió un error inesperado:", error)
else:
    print("El archivo se abrió correctamente.")
    contenido = archivo.read()
    print("Contenido leído:")
    print(contenido)
    archivo.close()
finally:
    print("Proceso finalizado.")
```
