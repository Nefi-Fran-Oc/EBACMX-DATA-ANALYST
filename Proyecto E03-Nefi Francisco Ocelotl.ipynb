```python
# FunciÃ³n de limpieza general
def clean_df(df):
    # 1. Estandarizar nombres de columnas
    df.columns = df.columns.str.strip().str.upper()
    
    # 2. Quitar espacios en datos tipo string
    for col in df.select_dtypes(include=["object"]):
        df[col] = df[col].astype(str).str.strip()

    # 3. Eliminar duplicados
    df = df.drop_duplicates()

    return df

# Aplicar la funciÃ³n
dim_category = clean_df(dim_category)
dim_product = clean_df(dim_product)
dim_segment = clean_df(dim_segment)
dim_calendar = clean_df(dim_calendar)
fact_sales = clean_df(fact_sales)
```

```python
#Para CATEGORY convertimos el ID a enteros
dim_category["ID_CATEGORY"] = pd.to_numeric(dim_category["ID_CATEGORY"], errors="coerce")
dim_category = dim_category.dropna(subset=["ID_CATEGORY"])
```

```python
#Para PRODUCT
# Convertir ITEM a string para unir con SALES
dim_product["ITEM"] = dim_product["ITEM"].astype(str).str.replace(".0","")

# Convertir ID_ITEM a numÃ©rico
dim_product["ID_ITEM"] = pd.to_numeric(dim_product["ID_ITEM"], errors="coerce")

# Rellenar NO DEFINIDO con NaN
dim_product.replace("NO DEFINIDO", np.nan, inplace=True)

#Convertir CATEGORY A string para unir con SALES
dim_product['CATEGORY'] = dim_product['CATEGORY'].astype(str)
```

```python
#Para SEGMENT se hace la relaciÃ³n con PRODUCT
dim_segment["ID_SEGMENT"] = pd.to_numeric(dim_segment["ID_SEGMENT"], errors="coerce")
```

```python
#Para CALENDAR se ajustan los formatos de fecha para que se puedan utilizar correctamente
dim_calendar["DATE"] = pd.to_datetime(dim_calendar["DATE"], format='mixed', dayfirst=True)

dim_calendar["WEEK_NUMBER"] = pd.to_numeric(dim_calendar["WEEK_NUMBER"], errors="coerce")
dim_calendar["YEAR"] = pd.to_numeric(dim_calendar["YEAR"], errors="coerce")
```

```python
#Para SALES
# ITEM_CODE a string para unir con PRODUCT
fact_sales["ITEM_CODE"] = fact_sales["ITEM_CODE"].astype(str).str.replace(".0","")

# Convertir columnas numÃ©ricas
num_cols = ["TOTAL_UNIT_SALES", "TOTAL_VALUE_SALES", "TOTAL_UNIT_AVG_WEEKLY_SALES"]

for col in num_cols:
    fact_sales[col] = pd.to_numeric(fact_sales[col], errors="coerce")
#Convertir CATEGORY A string para unir con PRODUCT
dim_product['CATEGORY'] = dim_product['CATEGORY'].astype(str)
```

```python
#UniÃ³n de tablas
#SALES con Product
sales_prod = fact_sales.merge(
    dim_product,
    how="left",
    left_on="ITEM_CODE",
    right_on="ITEM"
)
sales_prod_cat = sales_prod.merge(
    dim_category,
    how="left",
    left_on="CATEGORY",
    right_on="CATEGORY"
)
final_df = sales_prod_cat.merge(
    dim_calendar,
    how="left",
    on="WEEK"
)
```
