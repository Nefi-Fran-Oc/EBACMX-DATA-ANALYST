```python
#Importación de librerías
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, avg, year, to_timestamp, to_date
import os
```

```python
#Creación de sesión
spark = SparkSession \
    .builder \
    .appName("ETL_Housing") \
    .getOrCreate()
```

```python
#Carga de datos
path_archivo = '/Users/Nefi/OneDrive/Documentos/09-EBAC/EBACMX-DATA-ANALYST/Referencias/M54housing_clean.csv'
df = spark.read.csv(path_archivo, header=True, inferSchema=True)
```

```python
# Convertir a tipo fecha correctamente
df = df.withColumn(
    "date",
    to_date(col("date"), "yyyy-MM-dd")
)
```

```python
# Crear columna año
df = df.withColumn(
    "year",
    year(col("date"))
)
```

```python
# TABLA 1

tabla_1 = df.groupBy("zipcode", "year").agg(
    avg("price").alias("precio_promedio"),
    avg("sqft_living").alias("tamano_promedio"),
    avg("bedrooms").alias("habitaciones_promedio")
)

tabla_1.show()
```

```python
# TABLA 2

tabla_2 = df.groupBy("zipcode", "year").agg(
    avg("price").alias("precio_promedio")
)

tabla_2.show()
```

```python
#Exportado en csv
tabla_1.coalesce(1).write.mode("overwrite").option("header", True).csv("tabla_1_csv")
tabla_2.coalesce(1).write.mode("overwrite").option("header", True).csv("tabla_2_csv")
```
