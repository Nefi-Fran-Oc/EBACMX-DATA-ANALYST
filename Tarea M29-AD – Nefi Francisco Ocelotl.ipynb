```python
#Ejercicio 1
import requests
from bs4 import BeautifulSoup

url = "https://example.com"

response = requests.get(url)

print("Código de estado:", response.status_code)
print("\nContenido del sitio:\n")
print(response.text)
```

```python
#Ejercicio 2

soup = BeautifulSoup(response.text, "html.parser")

# Extraer el título de la página
title = soup.find("h1")

# Extraer todos los párrafos
paragraphs = soup.find_all("p")

print("Título de la página:")
print(title.text)

print("\nPárrafos encontrados:")
for p in paragraphs:
    print("-", p.text)
```

```python
api_url = "https://jsonplaceholder.typicode.com/posts"

api_response = requests.get(api_url)

print("Código de estado:", api_response.status_code)

data = api_response.json()

print("\nPrimer elemento de la respuesta:")
print(data[0])
```

```python
titles = [post["title"] for post in data]

titles_sorted = sorted(titles)

print("Listado ordenado de títulos:\n")
for title in titles_sorted:
    print("-", title)
```
