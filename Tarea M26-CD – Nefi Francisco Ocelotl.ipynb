```python
#Importo desde el inicio las librerías que podría utilizar
import csv
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

cwd = os.getcwd()
print("Directorio: {0}".format(cwd))
```

```python
#Abro el documento en formato csv
file = open(r'C:\Height of Male and Female by Country 2022.csv')

csvreader = csv.reader(file)
```

```python
#Imprimir visualmente los header
header = []
header = next(csvreader)
header
```

```python
#Extraer los datos a una lista
rows = []
for row in csvreader:
    rows.append(row)
print(rows[:50])
```

```python
#Generamos el objeto alturas con todas las filas de la lista
alturas = np.array(rows)
```

```python
#Generar array solo de las alturas, tanto de hombres como de mujeres
alturas_hombres = alturas[:,2]
alturas_mujeres = alturas[:,3]

#Transformamos la información en números con decimales
alturas_hombres_cm = [float(numeric_string) for numeric_string in alturas_hombres]
alturas_mujeres_cm = [float(numeric_string) for numeric_string in alturas_mujeres]
```

```python
#Calculamos la media, mínima y máxima de nuestros datos
#Hombres
print(f"Media de altura en hombres: {np.min(alturas_hombres_cm):.2f}")
print(f"Media de altura en hombres: {np.mean(alturas_hombres_cm):.2f}")
print(f"Media de altura en hombres: {np.max(alturas_hombres_cm):.2f}")

#Mujeres
print(f"Media de altura en mu: {np.mean(alturas_hombres_cm):.2f}")
print(f"Media de altura en hombres: {np.mean(alturas_hombres_cm):.2f}")
print(f"Media de altura en hombres: {np.mean(alturas_hombres_cm):.2f}")
```
