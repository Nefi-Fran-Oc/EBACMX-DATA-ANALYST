```python
df_housing = pd.read_csv("Analista de datos M54 - kc_house_data.csv")

# Limpiar nombres de columnas
df_housing.columns = [col.strip().lower().replace(" ", "_") for col in df_housing.columns]

# Convertir tipos
df_housing['date'] = pd.to_datetime(df_housing['date'], errors='coerce')

cols_numericas = [
    'price','bedrooms','bathrooms','sqft_living','sqft_lot',
    'floors','waterfront','view','condition','grade',
    'sqft_above','sqft_basement','yr_built','yr_renovated',
    'lat','long','sqft_living15','sqft_lot15'
]

for col in cols_numericas:
    df_housing[col] = pd.to_numeric(df_housing[col], errors='coerce')

# Guardar limpio
df_housing.to_csv("M54housing_clean.csv", index=False)
```

```python
df_sales = pd.read_csv("Analista de datos M54 - supermarket_sales.csv")

# Limpiar columnas
df_sales.columns = [col.strip().lower().replace(" ", "_") for col in df_sales.columns]

# Convertir tipos
df_sales['date'] = pd.to_datetime(df_sales['date'], errors='coerce')

cols_sales = [
    'unit_price','quantity','tax_5%','total',
    'cogs','gross_margin_percentage','gross_income','rating'
]

for col in cols_sales:
    df_sales[col] = pd.to_numeric(df_sales[col], errors='coerce')

# Guardar limpio
df_sales.to_csv("M54sales_clean.csv", index=False)
```

```python
df_wine = pd.read_csv("Analista de datos M54 - wine-clustering.csv")

# Limpiar columnas
df_wine.columns = [col.strip().lower().replace(" ", "_") for col in df_wine.columns]

# Convertir todo a numérico
for col in df_wine.columns:
    df_wine[col] = pd.to_numeric(df_wine[col], errors='coerce')

# Guardar limpio
df_wine.to_csv("M54wine_clean.csv", index=False)
```
