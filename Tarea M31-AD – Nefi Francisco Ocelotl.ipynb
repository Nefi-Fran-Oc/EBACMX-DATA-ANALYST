```python
#Lambda con zona horaria y capital
def info_pais(pais):
    datos = {
        'Argentina': ('UTC-3', 'Buenos Aires'),
        'Brazil': ('UTC-3', 'Brasilia'),
        'Spain': ('UTC+1', 'Madrid'),
        'Germany': ('UTC+1', 'Berlin'),
        'Mexico': ('UTC-6', 'Mexico City'),
        'France': ('UTC+1', 'Paris'),
        'Belgium': ('UTC+1', 'Brussels'),
        'Portugal': ('UTC+0', 'Lisbon'),
        'England': ('UTC+0', 'London')
    }
    return datos.get(pais, ('Desconocido', 'Desconocida'))

df[['Zona_Horaria', 'Capital']] = df['Nationality'].apply(
    lambda x: pd.Series(info_pais(x))
)

df[['Nationality', 'Zona_Horaria', 'Capital']].head()
```

```python
#Lambda para Candidato a balón de oro
df['Balon_de_Oro'] = df['Potential'].apply(
    lambda x: 'Candidato' if x > 85 else 'Next Year'
)

df[['Name', 'Potential', 'Balon_de_Oro']].head(10)
```
