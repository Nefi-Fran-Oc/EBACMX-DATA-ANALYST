```python
#Ejemplo de raíz cuadrada sin Lambda
def raiz_cuadrada(n):
    return n ** 0.5

print(raiz_cuadrada(25))
```

```python
#Ejemplo de raíz cuadrada con Lambda
raiz = lambda x: x ** 0.5
print(raiz(25))
```

```python
#Ejemplo para obtener el largo de cada palabra sin Map
texto = "Python es un gran lenguaje"
palabras = texto.split()

longitudes = []
for palabra in palabras:
    longitudes.append(len(palabra))

print(longitudes)
```

```python
#Ejemplo para obtener el largo de cada palabra con Map
texto = "Python es un gran lenguaje"
palabras = texto.split()

longitudes = list(map(lambda x: len(x), palabras))

print(longitudes)
```

```python
#Ejemplo para calcular el producto de una lista sin Reduce
numeros = [2, 3, 4, 5]

producto = 1
for n in numeros:
    producto *= n

print(producto)
```

```python
#Ejemplo para calcular el producto de una lista con Reduce
from functools import reduce

numeros = [2, 3, 4, 5]

producto = reduce(lambda x, y: x * y, numeros)

print(producto)
```

```python
#Ejemplo para encontrar palabras que contengan mayúsculas o números sin filter
palabras = ["Hola", "mundo", "Python3", "data", "ANÁLISIS", "curso2025"]

resultado = []
for palabra in palabras:
    if any(char.isupper() or char.isdigit() for char in palabra):
        resultado.append(palabra)

print(resultado)
```

```python
#Ejemplo para encontrar palabras que contengan mayúsculas o números sin filter
palabras = ["Hola", "mundo", "Python3", "data", "ANÁLISIS", "curso2025"]

resultado = list(filter(lambda x: any(c.isupper() or c.isdigit() for c in x), palabras))

print(resultado)
```
